#### 4. Reading a Table on a particular page of a PDF File.

```python
file1 = "./data/data.pdf"
```

```python
# To read table in first page of PDF file
table1 = tabula.read_pdf(file1 ,pages="all")
# To read tables in secord page of PDF file
#table2 = tabula.read_pdf(file1 ,pages=2)
print(table1[0])
print(table1[1])
```

```python
file2 = "./data/data3.pdf"
```

#### 5. What if there are multiple tables on the same page of a PDF file?


##### To read multiple tables we need to add extra parameter
##### multiple_tables = True -> Read multiple tables as independent tables
##### multiple_tables = False -> Read multiple tables as single table

#### 5.1. Reads Multiple Tables as Independent tables


```python
tables = tabula.read_pdf(file2 ,pages=1, multiple_tables=True)
print(tables[0])
print(tables[1])
```

#### 5.2 Reading Multiple Table as a Single Table

```python
tables = tabula.read_pdf(file2 ,pages=1,multiple_tables=False)
tables[0]
```

#### 6. Covert a PDF file directly to a CSV file
##### we can directly convert a PDF file containing tabular data directly to a CSV file using convert_into() method in tabula library.

#### 1. Converting tables in 1 page of PDF file to CSV

```python
tabula.convert_into("./data/data1.pdf", "./data/new_data1.csv",pages="1")
```

#### 2. Converting all table in PDF file to CSV

```python
tabula.convert_into("./data/data.pdf", "./data/new_data.csv",pages="all")
```

```python
tabula.convert_into("./data/data3.pdf","./data/new_data3.csv",pages="all")
```
